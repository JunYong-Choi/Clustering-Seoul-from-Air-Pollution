### ***대기Data 전처리***

```python
daegi_data.head(3)
```

```python
# 2020,2021 아황산가스농도(ppm) -> 미세먼지농도(㎍/㎥)
# 초미세먼지농도(㎍/㎥) 와 2022 아황산가스(ppm) -> 미세먼지(㎍/㎥),초미세먼지(㎍/㎥) 합치기
daegi_data['아황산가스농도(ppm)'] = daegi_data['아황산가스농도(ppm)'].fillna(daegi_data['아황산가스(ppm)'])
daegi_data['미세먼지농도(㎍/㎥)'] = daegi_data['미세먼지농도(㎍/㎥)'].fillna(daegi_data['미세먼지(㎍/㎥)'])
daegi_data['초미세먼지농도(㎍/㎥)'] = daegi_data['초미세먼지농도(㎍/㎥)'].fillna(daegi_data['초미세먼지(㎍/㎥)'])
daegi_data = daegi_data.iloc[:,:-3]
daegi_data
```

#### Complete

```python
clust_complete = linkage(daegi_3year, method='complete')

plt.figure(figsize=(13,6))
dendrogram(clust_complete)
plt.xticks(rotation=60, fontsize=15)
plt.show()
```

#### Average

```python
clust_avg = linkage(daegi_3year, method='average')

plt.figure(figsize=(14,6))
dendrogram(clust_avg)
plt.xticks(rotation=60, fontsize=15)
plt.show()
```

#### Ward

```python
clust_ward = linkage(daegi_3year, method='ward')

plt.figure(figsize=(14,6))
dendrogram(clust_ward)
plt.xticks(rotation=60, fontsize=15)
plt.show()
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=4).fit(daegi_3year)

daegi_3year['cluster_km']=km.labels_
daegi_3year['cluster_km']=daegi_3year['cluster_km'].astype('category')

daegi_3year.cluster_km.value_counts()
```

```python
# 적절한 군집화 개수 파악
wss=[]
for k in np.arange(1,20):
    km_k = KMeans(n_clusters=k).fit(daegi_3year)
    wss.append(km_k.inertia_)
```

```python
plt.plot(np.arange(1,20),wss)
plt.xlabel('K')
plt.ylabel('WSS')
plt.show()
# 3,4,5개가 적당하다고 파악
```

```python
daegi_3year2 = daegi_data2.groupby('측정소명')[daegi_data2.columns[2:]].mean().reset_index()
```

```python
from sklearn.decomposition import PCA # 주성분 분석
from sklearn.preprocessing import StandardScaler # 표준화
from pca import pca
model=pca(n_components=6, normalize=True) # 표준화 진행
result = model.fit_transform(daegi_3year2)
```
