# 3. Data Cleansing & Feature Engineering

```python
ha1000_data = pd.concat([ha1000_20, ha1000_21, ha1000_22], axis=0)
daegi_data = pd.concat([daegi_20, daegi_21, daegi_22], axis=0)
```

### ***하천Data 전처리***

```python
ha1000_data['측정소명'] = ha1000_data['측정소명'].apply(lambda x: x[:-3])
```

```python
ha1000_data.head(3)
```

```python
ha1000_data.columns[4:-2]
```

```python
f = ha1000_data[['날짜','측정소명']]
f = f.drop_duplicates().sort_values(by = ['날짜', '측정소명']).reset_index(drop=True)
```

```python
for column in ha1000_data.columns[4:-2]:
    t = ha1000_data.groupby(['날짜', '측정소명'])[column].agg([(column,"mean")]).reset_index()
    f = pd.concat([f, t[column]], axis = 1)

ha1000_data = f
```

```python
ha1000_data
```

### ***대기Data 전처리***

```python
daegi_data.head(3)
```

노량진측정소 : 동작구, 동작대로, 영등포구, 영등포로, 용산구, 한강대로  
선유측정소   : 강서구, 공항대로, 마포구, 마포아트센터, 양천구, 영등포구, 영등포로  
안양천측정소 : 구로구, 궁동, 금천구, 시흥대로  
탄천측정소   : 강남구, 강남대로, 도산대로, 세곡, 송파구, 올림픽공원  
중랑천측정소 : 동대문구, 중랑구, 홍릉로

```python
lst_노량진 = ['동작구', '동작대로', '영등포구', '영등포로','용산구', '한강대로']
lst_선유   = ['강서구', '공항대로', '마포구', '마포아트센터', '양천구', '영등포구', '영등포로']
lst_안양천 = ['구로구', '궁동', '금천구', '시흥대로']
lst_탄천   = ['강남구', '강남대로', '도산대로', '세곡', '송파구', '올림픽공원']
lst_중랑천 = ['동대문구', '중랑구', '홍릉로']
```

```python
daegi_노량진 = daegi_data.loc[daegi_data['측정소명'].isin(lst_노량진)].reset_index(drop = True)
daegi_선유 = daegi_data.loc[daegi_data['측정소명'].isin(lst_선유)].reset_index(drop = True)
daegi_안양천 = daegi_data.loc[daegi_data['측정소명'].isin(lst_안양천)].reset_index(drop = True)
daegi_탄천 = daegi_data.loc[daegi_data['측정소명'].isin(lst_탄천)].reset_index(drop = True)
daegi_중랑천 = daegi_data.loc[daegi_data['측정소명'].isin(lst_중랑천)].reset_index(drop = True)
```

```python
노량진 = daegi_노량진[['측정일시']].drop_duplicates().sort_values(by = ['측정일시']).reset_index(drop=True)
선유 = daegi_선유[['측정일시']].drop_duplicates().sort_values(by = ['측정일시']).reset_index(drop=True)
안양천 = daegi_안양천[['측정일시']].drop_duplicates().sort_values(by = ['측정일시']).reset_index(drop=True)
탄천 = daegi_탄천[['측정일시']].drop_duplicates().sort_values(by = ['측정일시']).reset_index(drop=True)
중랑천 = daegi_중랑천[['측정일시']].drop_duplicates().sort_values(by = ['측정일시']).reset_index(drop=True)
```

```python
노량진['측정소명'] = '노량진'
선유['측정소명'] = '선유'
안양천['측정소명'] = '안양천'
탄천['측정소명'] = '탄천'
중랑천['측정소명'] = '중랑천'
```

```python
column_lst = daegi_data.columns[2:]

for column in column_lst:
    t = daegi_노량진.groupby(['측정일시'])[column].agg([(column,"mean")]).reset_index()
    노량진 = pd.concat([노량진, t[column]], axis = 1)

for column in column_lst:
    t = daegi_선유.groupby(['측정일시'])[column].agg([(column,"mean")]).reset_index()
    선유 = pd.concat([선유, t[column]], axis = 1)
    
for column in column_lst:
    t = daegi_안양천.groupby(['측정일시'])[column].agg([(column,"mean")]).reset_index()
    안양천 = pd.concat([안양천, t[column]], axis = 1)

for column in column_lst:
    t = daegi_탄천.groupby(['측정일시'])[column].agg([(column,"mean")]).reset_index()
    탄천 = pd.concat([탄천, t[column]], axis = 1)

for column in column_lst:
    t = daegi_중랑천.groupby(['측정일시'])[column].agg([(column,"mean")]).reset_index()
    중랑천 = pd.concat([중랑천, t[column]], axis = 1)
```

```python
daegi_data = pd.concat([노량진, 선유, 안양천, 탄천, 중랑천], axis = 0, join='inner').sort_values(by = ['측정일시','측정소명']).reset_index(drop = True)
```

```python
data = pd.merge(ha1000_data, daegi_data, how='outer', left_on=['날짜', '측정소명'], right_on=['측정일시', '측정소명'])
```

```python
data.loc[data['날짜'].isnull()].측정일시.unique()
```

일단 누락된 날짜는 전부 제거하겠습니다.

```python
idx = data[data['날짜'].isnull()].index
data.drop(idx, inplace = True)    # 날짜 null 값 삭제
```

```python
data.drop(['측정일시'], axis = 1, inplace = True)  # 중복되는 날짜 삭제
```

```python
data['날짜'] = data['날짜'].apply(lambda x: int(x))
data['날짜'] = pd.to_datetime(data['날짜'], format = '%Y%m%d') # 날짜 datetime 형식으로 변환
```
